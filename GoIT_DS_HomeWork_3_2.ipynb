```python
import numpy as np
from random import triangular, random, choice
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coords = [(0, 0), (10,30), (20,0)]
```

```python
# Случайная точка в треугольнике
def random_point(coords):
    print(coords)
    x, y = random(), random()
    q = abs(x - y)
    s, t, u = q, 0.5 * (x + y - q), 1 - 0.5 * (q + x + y)
    return (
        s * coords[0][0] + t * coords[1][0] + u * coords[2][0],
        s * coords[0][1] + t * coords[1][1] + u * coords[2][1],
    )
```

```python
# проверка
pt1 = (0, 0)
pt2 = (10, 25)
pt3 = (25, 2)
triangle = [pt1, pt2, pt3]
points = [random_point([pt1, pt2, pt3]) for _ in range(10000)]

x, y = zip(*points)
plt.scatter(x, y, s=0.1)
plt.show()
```

```python
# случайная точка внутри
rand_point = random_point(triangle)
rand_point
```

```python
# случайная вершина
random_cortex = choice(triangle)
random_cortex
```

```python
# точка, равноотстоящая от первой выбранной точки и от выбранной вершины
mean_point = ((rand_point[0] + random_cortex[0])/2, (rand_point[1] + random_cortex[1])/2)
mean_point
```

```python
t = [p for p in triangle if p != random_cortex]
t.append(mean_point)
t
```

```python
def single_round():
    global triangle
    global points
    global rand_point

    random_cortex = choice(triangle)
    mean_point = ((rand_point[0] + random_cortex[0])/2, (rand_point[1] + random_cortex[1])/2)
    points.append(mean_point)
    # print(random_cortex)
    # print(triangle)
    triangle = [p for p in triangle if p != random_cortex]
    triangle.append(mean_point)
    # print(triangle)
    rand_point = mean_point
    # return triangle, points
```
