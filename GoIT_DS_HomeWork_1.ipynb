```python
def get_polynom(coords):
    coords = np.matrix(coords)

    n = coords.shape[0] - 1
    x = coords[:,0]
    b = coords[:,1]
    a = [[x_n.item()**nn for nn in range(n+1)] for x_n in x]
    return np.linalg.solve(a, b)
```

```python
coords = [[1,11], [2,22], [3,33], [4,44]]
```

```python
get_polynom(coords)
```
